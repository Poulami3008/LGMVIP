```python
##importing required libraries
import cv2  
import matplotlib.pyplot as plt
```

```python
##plotting the original image(in RGB format)
fig = plt.figure(figsize=(10,7))
image=cv2.imread('Panda.jpg')
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image, aspect='auto')
plt.show()
```

```python
##creating grayscale image from original image
fig = plt.figure(figsize=(10,7))
Gray_scl_image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(Gray_scl_image, cmap='gray', aspect='auto')
plt.show()
```

```python
##creating negative image from grayscale image
fig = plt.figure(figsize=(10,7))
Ngtv_img=cv2.bitwise_not(Gray_scl_image, cv2.COLOR_BGR2GRAY)
plt.imshow(Ngtv_img, cmap='gray', aspect='auto')
plt.show()
```

```python
##creating blurred image from negative image
fig = plt.figure(figsize=(10,7))
Blr_img=cv2.GaussianBlur(Ngtv_img, (21,21),0)
plt.imshow(Blr_img, cmap='gray', aspect='auto')
plt.show()
```

```python
##creating pencil sketch with the help of inverted blur and grayscale image
fig = plt.figure(figsize=(10,7))
Inverted_Blr_img=255-Blr_img
Pencil_Sketch=cv2.divide(Gray_scl_image, Inverted_Blr_img, scale=256)
plt.imshow(Pencil_Sketch, cmap='gray', aspect='auto')
plt.show()
```

```python
##original image and pencil sketch are in one frame
fig = plt.figure(figsize=(20,7))

plt.subplot(1,2,1)
plt.imshow(image,  aspect='auto')


plt.subplot(1,2,2)
plt.imshow(Pencil_Sketch, cmap='gray',  aspect='auto')

plt.show()
```
