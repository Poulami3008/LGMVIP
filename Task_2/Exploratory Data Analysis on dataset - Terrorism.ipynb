## Exploratory Data Analysis 

```python
data.rename(columns={'iyear':'year', 'imonth':'month', 'iday':'day', 'country_txt':'country_name', 'region_txt':'region_name', 'provstate':'state', 'attacktype1_txt':'attack_type', 'target1':'target', 'targtype1_txt':'target_type', 'gname':'group', 'nkill':'killed', 'nwound':'wounded', 'natlty1_txt':'nationality', 'weaptype1_txt':'weapon_type'}, inplace=True)
```

```python
data=data[['year', 'month', 'day', 'country_name', 'region_name', 'state', 'city', 'latitude', 'longitude', 'location', 'attack_type', 'target', 'target_type', 'group', 'killed', 'wounded', 'nationality', 'weapon_type', 'motive']]

data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum() ##checking for null values
```

```python
## remove columns having null values greater or equal to 70%
Null_per=round((((data.isnull().sum())/(len(data)))*100),2)
round(Null_per.sort_values(ascending=False),2)
```

```python
data.drop(['motive', 'location'], axis=1, inplace=True)
data.head()
```

```python
data.shape
```

```python
## filling NaN values of column 'killed' and 'wounded' with their mean values respectively.
data['killed'].fillna(int(data['killed'].mean()), inplace=True)
data['wounded'].fillna(int(data['wounded'].mean()), inplace=True)
data.head()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data.corr(), cmap='Reds', annot=True)
plt.show()
```

## By analysing data we can conclude the following:
### 1. Most attacks happen in the year 2014 and least attacks happen in the year 1971.
### 2. Iraq is the most attacked country followed by Pakistan and Afganishan.
### 3. Middle East & North Africa are the most attacked region worldwide. Australasia & Oceania comes at the bottom in this list.
### 4. Both state and citywise Baghdad faces most attacks.
### 5. Worldwide most attacks happen through Bombing and Explosions followed by firearms.
### 6. If we see the number of attacks, civilians are targeted mostly.
### 7. Groupwise most attacks happen by Taliban folowed by Islamic State of Iraq and the Levant (ISIL). National Liberation Army of Colombia (ELN) comes at the bootom of this list.
### 8. In 2014 highest number of people killed and in 2015 highest number of people wounded. (see the graph)
### 9. Iraq is the country where people are killed and wounded the most. 
### 10. Regionwise Middle East & North Africa come first both in the category of 'killed' and 'wounded'.
### 11. Again Baghdad is the most affected city and state in terms of killing and wounded.
