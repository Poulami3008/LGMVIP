```python
## encoding the target column
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
Label = LE.fit_transform(iris['Species'])
Label
```

## Preparing data for model

```python
iris.drop('Species', axis =1, inplace = True)
iris['Species'] = Label
```

```python
iris.head()
```

```python
X = iris.drop(columns = ['Species'])
Y = iris['Species']
```

```python
X.head()
```

```python
Y.head()
```

## Model training

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(X, Y, test_size = 0.3, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion = "entropy",random_state = 0, max_leaf_nodes = 10)
clf.fit(x_train, y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy is'), accuracy_score(y_test,y_pred)*100
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
## fit the model for whole dataset
dt_clf=DecisionTreeClassifier()
dt_clf.fit(X,Y)
```

```python
from sklearn import tree
text_presentation=tree.export_text(dt_clf)
print(text_presentation)
```

```python
## plotting the decision tree
plt.figure(figsize=(18,12))
tree.plot_tree(dt_clf, filled=True, impurity=True)
```
